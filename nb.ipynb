```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib import patches
```

```python
senna_df = pd.read_csv('data-csv/qualifying-results_ayrton-senna.csv')
prost_df = pd.read_csv('data-csv/qualifying-results_alain-prost.csv')
mansell_df = pd.read_csv('data-csv/qualifying-results_nigel-mansell.csv')
piquet_df = pd.read_csv('data-csv/qualifying-results_nelson-piquet.csv')
berger_df = pd.read_csv('data-csv/qualifying-results_gerhard-berger.csv')
cecotto_df = pd.read_csv('data-csv/qualifying-results_johnny-cecotto.csv')
johansson_df = pd.read_csv('data-csv/qualifying-results_stefan-johansson.csv')
dumfries_df = pd.read_csv('data-csv/qualifying-results_johnny-dumfries.csv')
angelis_df = pd.read_csv('data-csv/qualifying-results_elio-de-angelis.csv')
nakajima_df = pd.read_csv('data-csv/qualifying-results_satoru-nakajima.csv')
```

```python
years_1988_1994 = [1988, 1989, 1990, 1991, 1992, 1993, 1994]
years_until_1991 = [1988, 1989, 1990, 1991]
years_wo_1994 = [1988, 1989, 1990, 1991, 1992, 1993]

senna_data_1988_1994 = senna_df[senna_df['year'].isin(years_1988_1994)]
senna_data_1988_1994 = senna_data_1988_1994.reset_index()
prost_data_1988_1991 = prost_df[prost_df['year'].isin(years_until_1991)]
prost_data_1993 = prost_df[prost_df['year'] == 1993]
mansell_data_1988_1992 = mansell_df[mansell_df['year'].isin(years_wo_1994)]
piquet_data_1988_1991 = piquet_df[piquet_df['year'].isin(years_until_1991)]
berger_data_1988_1993 = berger_df[berger_df['year'].isin(years_wo_1994)]
berger_data_1994 = berger_df[berger_df['year'] == 1994].head(3)
berger_data_1988_1994 = pd.concat([berger_data_1988_1993, berger_data_1994], axis=0)

# In order to avoid SettingWithCopyWarning
senna_data_1988_1994 = senna_data_1988_1994.copy()
prost_data_1988_1991 = prost_data_1988_1991.copy()
prost_data_1993 = prost_data_1993.copy()
mansell_data_1988_1992 = mansell_data_1988_1992.copy()
piquet_data_1988_1991 = piquet_data_1988_1991.copy()
berger_data_1988_1994 = berger_data_1988_1994.copy()

senna_data_1988_1994['year_race'] = senna_data_1988_1994['race'] + " (" + senna_data_1988_1994['year'].astype(str) + ')'
prost_data_1988_1991['year_race'] = prost_data_1988_1991['race'] + " (" + prost_data_1988_1991['year'].astype(str) + ')'
prost_data_1993['year_race'] = prost_data_1993['race'] + " (" + prost_data_1993['year'].astype(str) + ')'
mansell_data_1988_1992['year_race'] = mansell_data_1988_1992['race'] + " (" + mansell_data_1988_1992['year'].astype(str) + ')'
piquet_data_1988_1991['year_race'] = piquet_data_1988_1991['race'] + " (" + piquet_data_1988_1991['year'].astype(str) + ')'
berger_data_1988_1994['year_race'] = berger_data_1988_1994['race'] + " (" + berger_data_1988_1994['year'].astype(str) + ')'
```

```python
fig, ax = plt.subplots(figsize=(28,8))

plt.plot(senna_data_1988_1994['year_race'], senna_data_1988_1994['position'], marker='o', linestyle='-', color='#e91c0d', markersize=3, label='Ayrton Senna')
plt.plot(prost_data_1988_1991['year_race'], prost_data_1988_1991['position'], marker='o', linestyle='-', color='#0066CC', markersize=3, label='Alain Prost')
plt.plot(prost_data_1993['year_race'], prost_data_1993['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='')
plt.plot(mansell_data_1988_1992['year_race'], mansell_data_1988_1992['position'], marker='o', linestyle='-', color='#ffb700', markersize=3, label='Nigel Mansell')
plt.plot(piquet_data_1988_1991['year_race'], piquet_data_1988_1991['position'], marker='o', linestyle='-', color='#38812F', markersize=3, label='Nelson Piquet')
plt.plot(berger_data_1988_1994['year_race'], berger_data_1988_1994['position'], marker='o', linestyle='-', color='#5a1b1b', markersize=3, label='Gerhard Beger')

plt.title('Ayrton Senna vs. Other Drivers - Qualifying Results (1988-1994)', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)
plt.gca().invert_yaxis()

loses = senna_data_1988_1994[senna_data_1988_1994['position'] != 1].index

# for lose in loses:
#     plt.axvspan(lose-0.5, lose+0.5, facecolor='#f8b4b4', alpha=0.5)

# for lose in loses:
#     x = senna_data_1988_1994['year_race'][lose]
#     y = senna_data_1988_1994['position'][lose]
#     center = (x, y) 
#     circle = patches.Circle(center, radius=1, color='#fc6969', alpha=0.5)
#     ax.add_patch(circle)

start_index_williams = senna_data_1988_1994[senna_data_1988_1994['year'] >= 1994].index[0] - senna_data_1988_1994[senna_data_1988_1994['year'] == 1990].index[0]
end_index_williams = len(senna_data_1988_1994['year_race'])-1
plt.axvspan(start_index_williams-0.5, end_index_williams+0.5, facecolor='#b4d6f8', alpha=0.5)

plt.axvspan(-0.5, len(senna_data_1988_1994[senna_data_1988_1994['year'] == 1988])-0.5, facecolor='#f7f1d0', alpha=0.5)

start_index_wc = senna_data_1988_1994[senna_data_1988_1994['year'] >= 1991].index[0] - senna_data_1988_1994[senna_data_1988_1994['year'] == 1989].index[0]
end_index_wc = start_index_wc + len(senna_data_1988_1994[senna_data_1988_1994['year'] == 1991])
plt.axvspan(start_index_wc, end_index_wc-0.5, facecolor='#f7f1d0', alpha=0.5)

start_index_wc = senna_data_1988_1994[senna_data_1988_1994['year'] >= 1992].index[0] - senna_data_1988_1994[senna_data_1988_1994['year'] == 1989].index[0]
end_index_wc = start_index_wc + len(senna_data_1988_1994[senna_data_1988_1994['year'] == 1992])
plt.axvspan(start_index_wc, end_index_wc-0.5, facecolor='#f7f1d0', alpha=0.5)


plt.grid(axis='y')

legend_elements = [
    Patch(facecolor='#8BC1F7', edgecolor='#b4d6f8', label='Williams Racing Dominance'),
    Patch(facecolor='#f5e9a3', edgecolor='#f7f1d0', label='Senna World Championships'),
    # Patch(facecolor='#f8b4b4', edgecolor='#f8b4b4', label="Senna's defeats"),
]

legend = plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements, loc='lower left')

plt.tight_layout()
plt.show()
```

### Toleman

```python
senna_data_1984 = senna_df[senna_df['year'] == 1984]
cecotto_data_1984 = cecotto_df[cecotto_df['year'] == 1984]
johansson_data_1984 = johansson_df[johansson_df['year'] == 1984]

# In order to avoid SettingWithCopyWarning
senna_data_1984 = senna_data_1984.copy()
cecotto_data_1984 = cecotto_data_1984.copy()
johansson_data_1984 = johansson_data_1984.copy()

senna_data_1984['year_race'] = senna_data_1984['race'] + " (" + senna_data_1984['year'].astype(str) + ')'
cecotto_data_1984['year_race'] = cecotto_data_1984['race'] + " (" + cecotto_data_1984['year'].astype(str) + ')'
johansson_data_1984['year_race'] = johansson_data_1984['race'] + " (" + johansson_data_1984['year'].astype(str) + ')'

loses_toleman = []

for index, (senna_row, cecotto_row) in enumerate(zip(senna_data_1984.iterrows(), cecotto_data_1984.iterrows())):
    senna_position = senna_row[1]['position']
    cecotto_position = cecotto_row[1]['position']

    if senna_position > cecotto_position:
        loses_toleman.append(index)
```

```python
plt.figure(figsize=(10,5))

plt.plot(senna_data_1984['year_race'], senna_data_1984['position'], marker='o', linestyle='-', color='#e91c0d', markersize=3, label='Ayrton Senna')
plt.plot(cecotto_data_1984['year_race'], cecotto_data_1984['position'], marker='o', linestyle='-', color='#0066CC', markersize=3, label='Johnny Cecotto')
plt.plot(johansson_data_1984['year_race'], johansson_data_1984['position'], marker='o', linestyle='-', color='#ffb700', markersize=3, label='Stefan Johansson')

plt.title('Ayrton Senna vs. Toleman Teammates - Qualifying Results', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)
plt.gca().invert_yaxis()

for lose in loses_toleman:
    plt.axvspan(lose-0.5, lose+0.5, facecolor='#f8b4b4', alpha=0.5)

legend_elements = [
    Patch(facecolor='#f8b4b4', edgecolor='#f8b4b4', label="Senna's defeats"),
]

legend = plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements, loc='upper left')

plt.tight_layout()
plt.show()
```

### McLaren

```python
mclaren_years = [1988, 1989, 1990, 1991, 1992, 1993]

senna_data_1988_1993 = senna_df[senna_df['year'].isin(mclaren_years)]
senna_data_1992_1993 = senna_df[senna_df['year'].isin([1992, 1993])]
berger_data_1992_1993 = berger_df[berger_df['year'].isin([1992, 1993])]

# In order to avoid senna_data_1988_1993
senna_data_1988_1993 = senna_data_1988_1993.copy()
berger_data_1992_1993 = berger_data_1992_1993.copy()

senna_data_1988_1993['year_race'] = senna_data_1988_1993['race'] + " (" + senna_data_1988_1993['year'].astype(str) + ')'
berger_data_1992_1993['year_race'] = berger_data_1992_1993['race'] + " (" + berger_data_1992_1993['year'].astype(str) + ')'

loses_mclaren = []

for index, (senna_row, prost_row) in enumerate(zip(senna_data_1988_1993.iterrows(), prost_data_1988_1991.iterrows())):
    senna_position = senna_row[1]['position']
    prost_position = prost_row[1]['position']

    if senna_position > prost_position:
        loses_mclaren.append(index)

for index, (senna_row, berger_row) in enumerate(zip(senna_data_1992_1993.iterrows(), berger_data_1992_1993.iterrows())):
    senna_position = senna_row[1]['position']
    berger_position = berger_row[1]['position']

    if senna_position > berger_position:
        lose_index = index + len(prost_data_1988_1991)
        loses_mclaren.append(lose_index+1)
```

```python
plt.figure(figsize=(18,5))

plt.plot(senna_data_1988_1993['year_race'], senna_data_1988_1993['position'], marker='o', linestyle='-', color='#e91c0d', markersize=2, label='Ayrton Senna')
plt.plot(prost_data_1988_1991['year_race'], prost_data_1988_1991['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='Alain Prost')
plt.plot(berger_data_1992_1993['year_race'], berger_data_1992_1993['position'], marker='o', linestyle='-', color='#ffb700', markersize=2, label='Gerhard Berger')

plt.title('Ayrton Senna vs. McLaren Teammates - Qualifying Results', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)
plt.gca().invert_yaxis()

plt.grid(axis='y')

for lose in loses_mclaren:
    plt.axvspan(lose-0.5, lose+0.5, facecolor='#f8b4b4', alpha=0.5)

legend_elements = [
    Patch(facecolor='#f8b4b4', edgecolor='#f8b4b4', label="Senna's defeats"),
]

legend = plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements, loc='lower left')

plt.tight_layout()
plt.show()
```
