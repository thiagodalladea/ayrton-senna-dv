```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
senna_df = pd.read_csv('data-csv/qualifying-results_ayrton-senna.csv')
prost_df = pd.read_csv('data-csv/qualifying-results_alain-prost.csv')
mansell_df = pd.read_csv('data-csv/qualifying-results_nigel-mansell.csv')
piquet_df = pd.read_csv('data-csv/qualifying-results_nelson-piquet.csv')
schumacher_df = pd.read_csv('data-csv/qualifying-results_michael-schumacher.csv')

years = [1988, 1989, 1990, 1991, 1992, 1993, 1994]
years_until_1991 = [1988, 1989, 1990, 1991]
years_wo_1994 = [1988, 1989, 1990, 1991, 1992, 1993]

senna_data = senna_df[senna_df['year'].isin(years)]
prost_data = prost_df[prost_df['year'].isin(years_until_1991)]
prost_data_1993 = prost_df[prost_df['year'] == 1993]
mansell_data = mansell_df[mansell_df['year'].isin(years_wo_1994)]
piquet_data = piquet_df[piquet_df['year'].isin(years_until_1991)]
schumacher_data_wo_1994 = schumacher_df[schumacher_df['year'].isin(years_wo_1994)]
schumacher_data_1994 = schumacher_df[schumacher_df['year'] == 1994].head(3)
schumacher_data = pd.concat([schumacher_data_wo_1994, schumacher_data_1994], axis=0)

# In order to avoid SettingWithCopyWarning
senna_data = senna_data.copy()
prost_data = prost_data.copy()
prost_data_1993 = prost_data_1993.copy()
mansell_data = mansell_data.copy()
piquet_data = piquet_data.copy()
schumacher_data = schumacher_data.copy()

senna_data['year_race'] = senna_data['race'] + " (" + senna_data['year'].astype(str) + ')'
prost_data['year_race'] = prost_data['race'] + " (" + prost_data['year'].astype(str) + ')'
prost_data_1993['year_race'] = prost_data_1993['race'] + " (" + prost_data_1993['year'].astype(str) + ')'
mansell_data['year_race'] = mansell_data['race'] + " (" + mansell_data['year'].astype(str) + ')'
piquet_data['year_race'] = piquet_data['race'] + " (" + piquet_data['year'].astype(str) + ')'
schumacher_data['year_race'] = schumacher_data['race'] + " (" + schumacher_data['year'].astype(str) + ')'
```

```python
plt.figure(figsize=(18,6))

start_index = senna_data[senna_data['year'] <= 1992].index[0]
plt.axvspan(start_index+1, len(senna_data['year_race'])-1, facecolor='#8BC1F7', alpha=0.5)
plt.plot(senna_data['year_race'], senna_data['position'], marker='o', linestyle='-', color='#C9190B', markersize=2, label='Ayrton Senna')
plt.plot(prost_data['year_race'], prost_data['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='Alain Prost')
plt.plot(prost_data_1993['year_race'], prost_data_1993['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='')
plt.plot(mansell_data['year_race'], mansell_data['position'], marker='o', linestyle='-', color='#F4C145', markersize=2, label='Nigel Mansell')
plt.plot(piquet_data['year_race'], piquet_data['position'], marker='o', linestyle='-', color='#38812F', markersize=2, label='Nelson Piquet')
plt.plot(schumacher_data['year_race'], schumacher_data['position'], marker='o', linestyle='-', color='#EC7A08', markersize=2, label='Michael Schumacher')

plt.title('Ayrton Senna vs Other Drivers - Qualifying Results (1988-1994)', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.gca().invert_yaxis()
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)

plt.grid(False)

legend_elements = [
    Patch(facecolor='#8BC1F7', edgecolor='#8BC1F7', label='Williams Racing Dominance'),
]

plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements)

plt.tight_layout()
plt.show()
```
