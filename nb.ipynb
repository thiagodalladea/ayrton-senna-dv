### Toleman

```python
senna_data_1984 = senna_df[senna_df['year'] == 1984]
cecotto_data_1984 = cecotto_df[cecotto_df['year'] == 1984]
johansson_data_1984 = johansson_df[johansson_df['year'] == 1984]

# In order to avoid SettingWithCopyWarning
senna_data_1984 = senna_data_1984.copy()
cecotto_data_1984 = cecotto_data_1984.copy()
johansson_data_1984 = johansson_data_1984.copy()

senna_data_1984['year_race'] = senna_data_1984['race'] + " (" + senna_data_1984['year'].astype(str) + ')'
cecotto_data_1984['year_race'] = cecotto_data_1984['race'] + " (" + cecotto_data_1984['year'].astype(str) + ')'
johansson_data_1984['year_race'] = johansson_data_1984['race'] + " (" + johansson_data_1984['year'].astype(str) + ')'

loses_toleman = []

for index, (senna_row, cecotto_row) in enumerate(zip(senna_data_1984.iterrows(), cecotto_data_1984.iterrows())):
    senna_position = senna_row[1]['position']
    cecotto_position = cecotto_row[1]['position']

    if senna_position > cecotto_position:
        loses_toleman.append(index)
```

```python
plt.figure(figsize=(10,5))

plt.plot(senna_data_1984['year_race'], senna_data_1984['position'], marker='o', linestyle='-', color='#e91c0d', markersize=3, label='Ayrton Senna')
plt.plot(cecotto_data_1984['year_race'], cecotto_data_1984['position'], marker='o', linestyle='-', color='#0066CC', markersize=3, label='Johnny Cecotto')
plt.plot(johansson_data_1984['year_race'], johansson_data_1984['position'], marker='o', linestyle='-', color='#ffb700', markersize=3, label='Stefan Johansson')

plt.title('Ayrton Senna vs. Toleman Teammates - Qualifying Results', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)
plt.gca().invert_yaxis()

for lose in loses_toleman:
    plt.axvspan(lose-0.5, lose+0.5, facecolor='#f8b4b4', alpha=0.5)

legend_elements = [
    Patch(facecolor='#f8b4b4', edgecolor='#f8b4b4', label="Senna's defeats"),
]

legend = plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements, loc='upper left')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
mclaren_years = [1988, 1989, 1990, 1991, 1992, 1993]

senna_data_1988_1993 = senna_df[senna_df['year'].isin(mclaren_years)]
senna_data_1992_1993 = senna_df[senna_df['year'].isin([1992, 1993])]
berger_data_1992_1993 = berger_df[berger_df['year'].isin([1992, 1993])]

# In order to avoid senna_data_1988_1993
senna_data_1988_1993 = senna_data_1988_1993.copy()
berger_data_1992_1993 = berger_data_1992_1993.copy()

senna_data_1988_1993['year_race'] = senna_data_1988_1993['race'] + " (" + senna_data_1988_1993['year'].astype(str) + ')'
berger_data_1992_1993['year_race'] = berger_data_1992_1993['race'] + " (" + berger_data_1992_1993['year'].astype(str) + ')'

loses_mclaren = []

for index, (senna_row, prost_row) in enumerate(zip(senna_data_1988_1993.iterrows(), prost_data_1988_1991.iterrows())):
    senna_position = senna_row[1]['position']
    prost_position = prost_row[1]['position']

    if senna_position > prost_position:
        loses_mclaren.append(index)

for index, (senna_row, berger_row) in enumerate(zip(senna_data_1992_1993.iterrows(), berger_data_1992_1993.iterrows())):
    senna_position = senna_row[1]['position']
    berger_position = berger_row[1]['position']

    if senna_position > berger_position:
        lose_index = index + len(prost_data_1988_1991)
        loses_mclaren.append(lose_index+1)
```
