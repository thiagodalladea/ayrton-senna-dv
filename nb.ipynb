```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
senna_data = pd.read_csv('data-csv/qualifying-results_ayrton-senna.csv')
prost_data = pd.read_csv('data-csv/qualifying-results_alain-prost.csv')
prost_data_1993 = pd.read_csv('data-csv/qualifying-results_alain-prost-1993.csv')
mansell_data = pd.read_csv('data-csv/qualifying-results_nigel-mansell.csv')
piquet_data = pd.read_csv('data-csv/qualifying-results_nelson-piquet.csv')
schumacher_data = pd.read_csv('data-csv/qualifying-results_michael-schumacher.csv')

senna_data['year_race'] = senna_data['race'] + " (" + senna_data['year'].astype(str) + ')'
prost_data['year_race'] = prost_data['race'] + " (" + prost_data['year'].astype(str) + ')'
prost_data_1993['year_race'] = prost_data_1993['race'] + " (" + prost_data_1993['year'].astype(str) + ')'
mansell_data['year_race'] = mansell_data['race'] + " (" + mansell_data['year'].astype(str) + ')'
piquet_data['year_race'] = mansell_data['race'] + " (" + mansell_data['year'].astype(str) + ')'
schumacher_data['year_race'] = schumacher_data['race'] + " (" + schumacher_data['year'].astype(str) + ')'
```

```python
plt.figure(figsize=(18,6))
start_index = senna_data[senna_data['year'] >= 1992].index[0]
plt.axvspan(start_index, len(senna_data['year_race'])-1, facecolor='#8BC1F7', alpha=0.5)
plt.plot(senna_data['year_race'], senna_data['position'], marker='o', linestyle='-', color='#C9190B', markersize=2, label='Ayrton Senna')
plt.plot(prost_data['year_race'], prost_data['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='Alain Prost')
plt.plot(prost_data_1993['year_race'], prost_data_1993['position'], marker='o', linestyle='-', color='#0066CC', markersize=2, label='')
plt.plot(mansell_data['year_race'], mansell_data['position'], marker='o', linestyle='-', color='#F4C145', markersize=2, label='Nigel Mansell')
plt.plot(piquet_data['year_race'], piquet_data['position'], marker='o', linestyle='-', color='#38812F', markersize=2, label='Nelson Piquet')
plt.plot(schumacher_data['year_race'], schumacher_data['position'], marker='o', linestyle='-', color='#EC7A08', markersize=2, label='Michael Schumacher')

plt.title('Ayrton Senna vs Other Drivers - Qualifying Results (1988-1994)', fontsize=14)
plt.xlabel('Race (Year)', fontsize=12)
plt.ylabel('Qualifying Position', fontsize=12)
plt.gca().invert_yaxis()
plt.xticks(rotation=90, fontsize=10)
plt.yticks([1, 5, 10, 15, 20, 25], fontsize=10)

plt.grid(False)

legend_elements = [
    Patch(facecolor='#8BC1F7', edgecolor='#8BC1F7', label='Williams Racing Dominance'),
]

plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + legend_elements)

plt.tight_layout()
plt.show()
```
